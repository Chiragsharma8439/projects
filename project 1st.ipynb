### we use the titanic dataset and predict person can survive or not 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv('titanic.csv')
```

```python
df.head()
```

### Remove the null values 


```python
df.isnull().sum()
```

```python
df1 = df.drop(['Pclass','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis= 'columns')# these cloumns are not dependent 
#on the survival of a person
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df2 = df1.fillna(method = 'ffill')
df2
```

```python
df2.isnull()
```

```python
target = df2['Survived']
target
```

### convert the categorical value into dummy variable 

```python
dummy = pd.get_dummies(df2.Sex)
dummy
```

```python
merge = pd.concat([df2,dummy],axis='columns')
```

```python
merge
```

```python
final = merge.drop('Sex',axis='columns')
```

```python
final
```

```python
x= final.drop('Survived',axis='columns')
x
```

```python
y= final.Survived
y
```

### trained the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x,target)
```

### prediction

```python
model.predict([[2,25,72.52,1,0]])
```

```python
model.predict([[5,35,64.8,0,1]])
```

```python
model.score(x,target)
```
